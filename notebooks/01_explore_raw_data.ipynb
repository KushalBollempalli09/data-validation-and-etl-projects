```python
missing_in_orders = set(order_items["order_id"]) - set(orders["order_id"])
len(missing_in_orders)
```

```python
order_items.groupby("order_id").size().sort_values(ascending=False).head(10)
```

```python
date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
]

for c in date_cols:
    orders[c] = pd.to_datetime(orders[c], errors="coerce")
```

```python
orders[date_cols].isna().mean().sort_values(ascending=False)
```

```python
orders["delivery_delay_days"] = (
    (orders["order_delivered_customer_date"] - orders["order_estimated_delivery_date"])
    .dt.days
)
```

```python
orders["fulfillment_days"] = (
    (orders["order_delivered_customer_date"] - orders["order_purchase_timestamp"])
    .dt.days
)
```

```python
orders[["delivery_delay_days", "fulfillment_days"]].describe()
```

```python
orders["purchase_month"] = orders["order_purchase_timestamp"].dt.to_period("M").astype(str)
orders[["order_purchase_timestamp", "purchase_month"]].head()
```
